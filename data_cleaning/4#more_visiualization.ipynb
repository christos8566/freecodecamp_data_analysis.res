## Import Library

```
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

print("pandas: ", pd.__version__)
print("numpy: ", np.__version__)
print("seaborn: ", sns.__version__)
```

# Global API

```
x = np.arange(-10, 11)

print(x ** 2)
x
```

Matplotlib's default pyplot API has a global, MATLAB-style interface, as we've already seen:

```
plt.figure(figsize = [12, 6], facecolor = 'y', edgecolor = 'g', linewidth = 10, alpha = 1)
plt.title("My Nice Plot", fontsize = 10, loc='left', color = 'purple', fontstyle = 'italic', fontweight = 'bold')

plt.plot(x, x**2, linestyle = '--')
plt.plot(x, -1 * (x**2))
```

```
plt.figure(figsize = [12, 6])
plt.title("My Nice Plot")
plt.subplot(1, 2, 1, polar = False) # numpers of rows, columns, panel number
plt.plot(x, x**2)
plt.plot([0, 0, 0], [-10, 10, 100])
plt.legend(['X^2', 'Vertical LIne'])
plt.xlabel("x")
plt.ylabel("x Squared")

plt.subplot(1, 2, 2,fc='pink')
plt.plot(x, -1 * (x**2))
plt.plot([-10, 0, 10], [-50, -50, -50])
plt.legend(["-x^2", "Horizontal Line"], fontsize = 10, labelcolor = 'r', facecolor = 'yellow', edgecolor = 'gray')
plt.xlabel("x", labelpad = 1, fontsize = 10, c= 'r', fontstyle = 'italic')
plt.ylabel("X Squared")
```

# OOP Interface


```
fig, axes = plt.subplots(figsize = [12, 6], facecolor = 'pink')
```

```
axes.plot(x,(x**2), c='r', linewidth = 3, marker = 'o', markersize = 8, label = 'x^2')
axes.plot(x, -1 * (x**2), linestyle = '--', label = '-x^2')

axes.set_xlabel('x', fontweight = 'bold', fontstyle = 'italic', c = 'blue')
axes.set_ylabel('x Squared', fontweight = 'bold', fontstyle = 'italic', c = 'blue')
axes.set_title('My Nice Plot')

axes.legend()
fig
```

```
fig,axes = plt.subplots(figsize = [12, 6], facecolor = 'red')
axes.plot(x, x + 0, linestyle = 'solid', label = 'x + 0')
axes.plot(x, x + 1, linestyle = 'dashed', label = 'x + 1')
axes.plot(x, x + 2, linestyle = 'dashdot', label = 'x + 2')
axes.plot(x, x + 3, linestyle = 'dotted', label = 'x + 3');

axes.set_title("My Nice Plot")
```

```
fig,axes = plt.subplots(figsize = [12, 6], facecolor = 'yellow')
axes.plot(x, x + 0, '-og', label = 'solid green')
axes.plot(x, x + 1, '--c', label = 'dashed cyan')
axes.plot(x, x + 2, '-.b', label = 'dashdot blue')
axes.plot(x, x + 3, ':r', label = 'dotted red');

axes.set_title("My Nice Plot", c = 'green', fontsize = 15, fontstyle = 'italic', fontweight = 'bold')

axes.legend()
```

```
# Creating subplots with shared x-axis
fig, (ax1, ax2) = plt.subplots(2, sharex=True, squeeze=True)

# Plotting data on first subplot
ax1.plot([1, 2, 3], [4, 5, 6])
ax1.set_title('Subplot 1')

# Plotting data on second subplot
ax2.plot([1, 2, 3], [6, 5, 4])
ax2.set_title('Subplot 2')

plt.show()
```

```
# Creating subplots with independent x-axis
fig, (ax1, ax2) = plt.subplots(2, sharex=False)

# Plotting data on first subplot
ax1.plot([1, 2, 3], [4, 5, 6])
ax1.set_title('Subplot 1')

# Plotting data on second subplot
ax2.plot([4, 5, 6], [1, 2, 3])
ax2.set_title('Subplot 2')

plt.show()
```

```
plot_object = plt.subplots()

fig, ax = plot_object

ax.plot([1, 2, 3], [1, 2, 3])
plot_object
```

We can also define how many elements we want inside our figure. To do that we can set the `nrows` and `ncols` params.



```
plot_object = plt.subplots(nrows = 2, ncols = 2, figsize = [12, 6])
ax = plot_object

fig, ((ax1, ax2), (ax3, ax4)) = plot_object
plot_object
```

```
ax4.plot(np.random.randn(50), c = 'y')
ax1.plot(np.random.randn(50), c ='red', linestyle = '--')
ax2.plot(np.random.randn(50), c ='g', linestyle = '-.')
ax3.plot(np.random.randn(50), c = 'blue', marker = 'o', linewidth = 3)

fig
```

## The `subplot2grid` command

There is another way to make subplots using a grid-like format:



```
plt.figure(figsize = [12, 6])

ax1 = plt.subplot2grid((3, 3), (0, 0), colspan = 3)
ax2 = plt.subplot2grid((3, 3), (1, 0), colspan = 2)
ax3 = plt.subplot2grid((3, 3), (1, 2), rowspan = 2)
ax4 = plt.subplot2grid((3, 3), (2, 0))
ax5 = plt.subplot2grid((3, 3), (2, 1))
```

## Scatter Plots



```
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * (20 * np.random.rand(N))**2  # 0 to 15 point radii
```

```
plt.figure(figsize = [12, 6])
plt.subplot(1,1,1,facecolor='white')# default color
plt.scatter(x, y,s=area, c= colors, cmap= 'Spectral',alpha= .5)
plt.colorbar()
plt.show()
```

```
fig = plt.figure(figsize= (14, 6))
ax1 = fig.add_subplot(1,2,1, facecolor='lightpink',)
plt.scatter(x, y, s = area, c = colors, alpha = .5, cmap = 'Pastel1')
plt.colorbar()

ax2 = fig.add_subplot(1,2,2, facecolor= 'lightgrey')
plt.scatter(x, y, s = area, c = colors, alpha = .5, cmap = 'Pastel2')
plt.colorbar()

plt.show()
```

## Histograms

```
values = np.random.randn(1000)
```

```
plt.subplots(figsize=[16,9],facecolor='yellow')
plt.hist(values, bins=100, alpha=.7,histtype='bar', color = 'steelblue', edgecolor = 'green')
plt.xlim(xmin=-5, xmax=5)

plt.show()
```

```
fig.savefig('hist.png')
```

## KDE (kernel density estimation)

```
from scipy import stats

density = stats.kde.gaussian_kde(values)
density
```

```
plt.subplots(figsize=[14, 6])

values2 = np.linspace(min(values)-10, max(values)+10, 100)

plt.plot(values2, density(values2), color = '#FF700F')
plt.fill_between(values2, 0, density(values2), alpha=.5, color= '#FF700F')

plt.xlim(xmin=-5, xmax=5)
plt.show()
```

### Combine plot

```
plt.subplots(figsize=(14, 6))

plt.hist(values, bins=100, alpha=.8, density=True, histtype='bar', color='steelblue', edgecolor='green')

plt.plot(values2, density(values2),color='#FF7F00', linewidth=3)
plt.xlim(xmin=-5, xmax=5)

plt.show()
```

### Bar plot

```
Y = np.random.rand(1, 5)[0]
Y2 = np.random.rand(1, 5)[0]
```

```
plt.figure(figsize=(12, 4))

barWidth = .5
plt.bar(np.arange(len(Y)), Y, width=barWidth, color='#00b894')

plt.show()
```

Also can be stacked bars, and add a legend to the plot:



```
plt.figure(figsize=(12, 4))

barWidth= .5

plt.bar(np.arange(len(Y)), Y, width=barWidth, color='#00b894',label='label Y')
plt.bar(np.arange(len(Y2)), Y2, width=barWidth, color='#e17055', bottom = Y, label= 'label Y2')

plt.legend()
plt.show()
```

### Boxplot and outlier detection

```
values = np.concatenate([np.random.randn(10), np.array([10, 15, -10, -15])])
```

```
plt.figure(figsize=(12, 4))

plt.hist(values)
```

```
plt.figure(figsize=[12, 4])

plt.boxplot(values)
```
