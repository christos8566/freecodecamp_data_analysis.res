# Series creation


## Import libraries

Import libraries of numpy and pandas

```python
# import the numpy package under the name of np
import numpy as np

# import the pandas package under the name of pd
import pandas as pd

# print the pandas version and certifiguration
print(pd.__version__)
```

## Create an empty pandas series

```python
pd.Series()
```

## Given the X python list, convert it on a y pandas Series

```python
X = ['A','B','C']
print(X, type(X))

Y = pd.Series(X)
print(Y, type(Y))
```

## Given the X pandas Series, name it 'My letters'


```python
X = pd.Series(['A', 'B', 'C'])

X.name = 'My letters'
X
```

## Given the X pandas series, show its values

```python
X = pd.Series(['A', 'B', 'C'])

X.values
```

# Series Indexation

## Assign index names to the given X pandas Series

```python
X = pd.Series(['A', 'B', 'C'])
index_names = ['First', 'Second', 'Third']          # X.index = ['First', 'Second', 'Thrird']

X.index = index_names
X
```
